# Blog Post:  Movie Correlations

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
movie_ratings = pd.read_csv('movie_ratings.csv')
corr_matrix = movie_ratings.corr()
x=movie_ratings["tomato_meter"]
y=movie_ratings["casey"]

plt.figure(figsize=(10, 6))
plt.scatter(
    x, y,
    marker = "o"
)

text=corr_matrix.loc["tomato_meter", "casey"]
plt.title(f"$correlation={text: 0.3f}$")
plt.xlabel('Tomato Meter')
plt.ylabel('My Ratings')
plt.xlim(0, 100)
plt.ylim(0, 100)

plt.show()
```

```python
plt.figure(figsize = (10,6))
sn.heatmap(corr_matrix, annot=True)
```
